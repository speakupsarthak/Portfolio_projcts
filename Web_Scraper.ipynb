```python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue#:~:text=List%20of%20the%20largest%20companies%20%20%20,%20%207.8%25%20%2023%20more%20rows%20'
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
table = soup.find_all('table')[1]
table_title = table.find_all('th')
title = [t.text.strip() for t in table_title]
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(columns = title)
```

```python
row_raw_data = table.find_all('tr')
for i in row_raw_data[1:]:
    row_data = i.find_all('td')
    individual_row_data = [r.text.strip() for r in row_data]
    
    length = len(data)
    data.loc[length] = individual_row_data
```
